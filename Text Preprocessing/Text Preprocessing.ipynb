### Remove HTML Tag

```python
# making a function to remove html tags

import regex as re
def remove_HTML_tags(text):
    pattern = re.compile("<.*?>")
    return pattern.sub(r'', text)

text = df['review'][3]
#text = "<br>This is a HTML tag</br>"
remove_HTML_tags(text)
```

```python
# To remove html tags from each data points in an entire column

# df['review'] = df['review'].apply(remove_HTML_tags)

for x in df['review']:
    remove_HTML_tags(x)

df['review'][71]
```

### Remove URL's

```python
import regex as re

def remove_urls(text):
    pattern = re.compile(r"https?://\S+|www\.\S+")
    return pattern.sub(r'', text).strip()

text1 = "Check out my instagram https://www.instagram.com/imsohamofc/qwerty"
remove_urls(text1)
text2 = "Check out my instagram http://www.instagram.com/imsohamofc/qwerty"
remove_urls(text2)
text3 = "Check out my instagram www.instagram.com/imsohamofc/qwerty"
remove_urls(text3)
text4 = "Check out my instagram https://www.instagram.com/imsohamofc/qwerty and also my facebook www.facebook.com/imsohamofc/qwerty"
remove_urls(text4)
```

### Removing Punctuation

```python
import string, time

exclude = string.punctuation

# Method 1
# def remove_punctuation(text):
#     for char in exclude:
#         text = text.replace(char, '')
#     return text

# Method 2 This method is much faster than the previous one
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', exclude))

text = "Hello, my name is Soham Mirashi. How do you do?"

remove_punctuation(text)
```

### Chat word treatment

```python
slang_dict = {}

with open("slang.txt", "r") as f:
    for line in f:
        line = line.strip()
        
        if "=" in line:
            key, value = line.split("=", 1)
            slang_dict[key.lower()] = value.lower()

print(slang_dict)
```

```python
def slang_conversion(text):
    words = text.split()
    return " ".join([slang_dict.get(word.lower(), word) for word in words])

slang_conversion("LOL that was sick!!")
slang_conversion("LO that was sick!!")
slang_conversion("FYI New Delhi is the capital of India")
```
